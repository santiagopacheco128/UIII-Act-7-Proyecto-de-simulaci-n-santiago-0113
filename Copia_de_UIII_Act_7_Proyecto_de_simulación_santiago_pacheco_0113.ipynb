```python
print("🎬 Sistema Inteligente de Actor y Director 🎬")

while True:

    print("\n====== MENÚ PRINCIPAL ======")
    print("1. Registrar actor")
    print("2. Revisar actuación")
    print("3. Evaluar casting")
    print("4. Revisar experiencia")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    # OPCIÓN 1
    if opcion == "1":
        nombre = input("Nombre del actor: ")
        edad = int(input("Edad: "))

        if edad >= 18:
            print(f"✅ Actor {nombre} registrado correctamente")
        else:
            print(f"⚠️ {nombre} es menor de edad")

    # OPCIÓN 2
    elif opcion == "2":
        nivel = input("Nivel de actuación (excelente/bueno/malo): ")

        if nivel == "excelente":
            print("🌟 El director quedó impresionado")
        elif nivel == "bueno":
            print("👍 El director está satisfecho")
        elif nivel == "malo":
            print("❌ Necesita más práctica")
        else:
            print("⚠️ Nivel no reconocido")

    # OPCIÓN 3
    elif opcion == "3":
        casting = int(input("Calificación del casting (0-10): "))

        if casting >= 9:
            print("🏆 Papel principal obtenido")
        elif casting >= 7:
            print("🎭 Papel secundario asignado")
        elif casting >= 5:
            print("🙂 Puede intentarlo nuevamente")
        else:
            print("❌ No fue seleccionado")

    # OPCIÓN 4
    elif opcion == "4":
        experiencia = int(input("Años de experiencia: "))

        if experiencia >= 10:
            print("🎬 Actor profesional")
        elif experiencia >= 5:
            print("🎥 Actor con experiencia")
        elif experiencia >= 1:
            print("📚 Actor principiante")
        else:
            print("👶 Nuevo en la actuación")

    # OPCIÓN 5
    elif opcion == "5":
        print("👋 Programa finalizado")
        print("Programa realizado por Santiago Pacheco NL 47")
        break

    # OPCIÓN INVÁLIDA
    else:
        print("⚠️ Opción inválida")
```
