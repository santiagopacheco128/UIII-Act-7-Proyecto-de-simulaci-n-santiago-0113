```python
print("Sistema Inteligente de Actor y Director")

while True:

    print("\n--- MENÚ ---")
    print("1. Registrar actor")
    print("2. Revisar actuación")
    print("3. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        nombre = input("Nombre del actor: ")
        edad = int(input("Edad: "))

        if edad >= 18:
            print("Actor registrado correctamente")
        else:
            print("Actor menor de edad")

    elif opcion == "2":
        nivel = input("Nivel de actuación (bueno/malo): ")

        if nivel == "bueno":
            print("El director está satisfecho")
        elif nivel == "malo":
            print("Necesita más práctica")
        else:
            print("Nivel no reconocido")

    elif opcion == "3":
        print("Programa finalizado")
        break

    else:
        print("Opción inválida")
```
